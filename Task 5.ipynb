```python


from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load small pretrained model
model_name = "distilgpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

print("Mental Health Chatbot Ready (type 'exit' to stop)\n")

def chat(user_input):
    
    # Simple safety response
    if any(word in user_input.lower() for word in ["suicide", "end my life", "die"]):
        return "I'm really sorry you're feeling this way. Please talk to a trusted person or mental health professional."

    input_text = f"User feels: {user_input}. Bot responds in a kind and supportive way:"
    
    inputs = tokenizer.encode(input_text, return_tensors="pt")
    
    output = model.generate(
        inputs,
        max_length=80,
        temperature=0.8,
        do_sample=True
    )
    
    return tokenizer.decode(output[0], skip_special_tokens=True)

# Chat loop
while True:
    user = input("You: ")
    
    if user.lower() == "exit":
        print("Bot: Take care 💙")
        break
    
    print("Bot:", chat(user))
```
