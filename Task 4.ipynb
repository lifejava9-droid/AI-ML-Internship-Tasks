```python


import os
from openai import OpenAI

# Set your API key here
os.environ["OPENAI_API_KEY"] = ""
client = OpenAI()

def chat(user_input):
    
    # Simple safety check
    danger_words = ["chest pain", "can't breathe", "suicide", "overdose", "stroke"]
    
    if any(word in user_input.lower() for word in danger_words):
        return "⚠️ This may be serious. Please seek immediate medical help or contact a doctor."

    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You are a helpful medical assistant. Give simple, safe, general health advice. Do not diagnose or give exact medicine doses."},
                {"role": "user", "content": user_input}
            ]
        )
        return response.choices[0].message.content

    except:
        return "Error: Check API key or internet connection."

# Chat loop
print("Health Chatbot Ready (type 'exit' to stop)\n")

while True:
    user = input("You: ")
    if user.lower() == "exit":
        print("Bot: Take care! 👋")
        break
    print("Bot:", chat(user))
```
